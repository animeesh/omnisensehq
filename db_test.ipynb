```python
import sqlite3

# Connect to the existing database
conn = sqlite3.connect("detections_yolov5.db")
cursor = conn.cursor()

# Fetch all detections
cursor.execute("SELECT * FROM detections")
rows = cursor.fetchall()

# Print results
print("Detections:")
print("Frame | Time (s) | Label | Conf | X | Y | W | H")
for row in rows:
    frame_id, timestamp, label, confidence, x, y, w, h = row
    print(f"{frame_id:5} | {timestamp:8.2f} | {label:10} | {confidence:.2f} | {x:3} | {y:3} | {w:3} | {h:3}")

conn.close()
```

```python
import requests
import json

api = "05d3b05f0bcd4116aa9110249252205"
import requests

API_KEY = "your_api_key"  # Replace with your actual API key
CITY = "Mumbai"
URL = f"http://api.worldweatheronline.com/premium/v1/weather.ashx"

params = {
    "key": api,
    "q": CITY,
    "format": "json",
    "num_of_days": 1,
    "tp": 1,  # hourly interval
    "cc": "yes"  # current conditions
}

response = requests.get(URL, params=params)
data = response.json()

# Extract current weather
current = data['data']['current_condition'][0]
temp_C = current['temp_C']
weather_desc = current['weatherDesc'][0]['value']
humidity = current['humidity']
wind_speed = current['windspeedKmph']

print(f"📍 City: {CITY}")
print(f"🌡️ Temperature: {temp_C}°C")
print(f"🌤️ Weather: {weather_desc}")
print(f"💧 Humidity: {humidity}%")
print(f"🌬️ Wind Speed: {wind_speed} km/h")
```

```python
import requests
import json
from datetime import datetime

api = "05d3b05f0bcd4116aa9110249252205"
CITY = "Mumbai"
URL = f"http://api.worldweatheronline.com/premium/v1/weather.ashx"

params = {
    "key": api,
    "q": CITY,
    "format": "json",
    "num_of_days": 0,
    "cc": "yes"  # current conditions
}

response = requests.get(URL, params=params)
data = response.json()

# Extract current weather
current = data['data']['current_condition'][0]
temp_C = current['temp_C']
weather_desc = current['weatherDesc'][0]['value']
now = datetime.now()
date_str = now.strftime("%Y-%m-%d")
time_str = now.strftime("%H:%M:%S")

print(f"📅 Date: {date_str}")
print(f"⏰ Time: {time_str}")
print(f"🌡️ Temperature: {temp_C}°C")
print(f"🌤️ Weather: {weather_desc}")

```
